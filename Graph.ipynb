```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.interpolate import make_interp_spline

# Sample data points (Change as needed)
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1, 6.3, 7.2, 8.1])

# 1. Linear Fit (y = ax + b)
linear_fit = np.polyfit(x, y, 1)
linear_fn = np.poly1d(linear_fit)

# 2. Polynomial Fit (Quadratic y = ax^2 + bx + c)
poly_fit = np.polyfit(x, y, 2)
poly_fn = np.poly1d(poly_fit)

# 3. Cubic Spline Interpolation
x_smooth = np.linspace(x.min(), x.max(), 100)
spline_fn = make_interp_spline(x, y, k=3)  # k=3 for cubic

# Plot the original data
plt.scatter(x, y, color='red', label="Data Points")

# Plot the different fits
plt.plot(x, linear_fn(x), label="Linear Fit", linestyle="--")
plt.plot(x, poly_fn(x), label="Quadratic Fit", linestyle="-.")
plt.plot(x_smooth, spline_fn(x_smooth), label="Cubic Spline", linestyle=":")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Best Fit Graphs")
plt.legend()
plt.grid()
plt.show()
```
